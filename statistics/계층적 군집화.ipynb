### 군집화
- 많은 샘플을 소수의 군집으로 묶어 각 군집의 특성을 파악하여 데이터의 특성을 이해
- 군집의 특성을 바탕으로 각 군집에 속하는 샘플들에 대한 세분화된 의사결정을 수행
#### 거리와 유사도
- 두 샘플이 유사하다 = 두 샘플 간 유사도가 높다 = 두 샘플 간 거리가 짧다
- 수치형 변수에 대해 정의되어 있으므로, 문자를 숫자로 바꿔주는 작업이 선행
#### 더미화
- 범주형 변수가 특정 값을 취하는지 여부를 나타내는 더비 변수 생성
- pandas.get_dummies : drop_first : 첫 번째 더미 변수를 제거할지 여부
- 숫자로 표현된 변수를 더미화 하려면, 반드시 str타입으로 변경
#### 유사도 척도
- 유클리디안 거리 : 가장 흔하게, 직선
- 맨하탄 거리 : 정수형 데이터(리커트 척도)에 적합, 수직/수평으로 이동한 거리 합
- 코사인 유사도 : 스케일을 고려하지 않고 방향 유사도를 측정(ex 추천시스템)
- 매칭 유사도 : 이진형 데이터에 적합한 유사도 척도로 전체 특징 중 일치하는 비율 고려
- 자카드 유사도 : 둘 중 하나라도 1을 가지는 특징 중 일치하는 비율 고려, 이진형에 적합

### 계층적 군집화
- 개별 샘플을 군집으로 간주, 가장 가까운 두 군집을 순차적으로 묶는 방식
- 최단 연결법 : 이상치에 민감, 계산량이 많은 편
- 최장 연결법 : 이상치에 민감, 계산량이 많은 편
- 평균 연결법 : 이상치에 둔감, 계산량이 많은 편
- 중심 연결법 : 이상치에 둔감, 계산량 적은 편
- 와드 연결법 : 이상치에 매우 둔감, 계산량 매우 많은 편, 군집 크기 비슷하게 만듦
- 덴드로그램 : 계측 군집화 과정을 트리 형태로 보여주는 그래프, 샘플 수가 많으면 해석이 불가능할 정도로 복잡
- 장점
 - 덴드로그램으로 군집화 과정확인 가능
 - 거리/유사도 행렬만 있으면 군집화 가능
 - 다양한 거리 척도 활용 가능
 - 수행할 때마다 같은 결과를 냄(임의성X)
- 단점
 - 상대적으로 많은 계산량
 - 군집 개수 설정에 제약 존재
- sklearn.cluster.AgglomerativeClustering
 - n_clusters : 군집화 개수
 - affinity : 거리 척도, linkage가 ward면 Euclidean만 가능
 - precomputed : 거리/유사도 행렬을 입력으로 하는 경우에 설정하는 값
 - linkage : 군집 간 거리
- pandas.crosstab : 거래 데이터를 교차 테이블 형태로 변환하는데 사용하는 함수

### 고객 특성에 따른 군집화

```python
import pandas as pd
df = pd.read_csv("Telco_customer_info.csv")
df.head()
```

```python
df.set_index('customerID', inplace = True)
```

```python
df.head()
```

```python
df = pd.get_dummies(df, drop_first = True)
df.head()
```

```python
df.columns
```

```python
from sklearn.cluster import AgglomerativeClustering
clusters = AgglomerativeClustering(n_clusters = 3,
                                   affinity = 'euclidean',
                                   linkage = 'ward').fit(df) 
```

```python
df['군집정보'] = clusters.labels_
```

```python
df['군집정보'].head()
```

```python
df.groupby(['군집정보'])[['MonthlyCharges', 'TotalCharges']].mean()
```

```python
df.groupby(['군집정보'])[['SeniorCitizen', 'StreamingTV_Yes']].mean()
```

### 구매기록 기준 고객 군집화

```python
df = pd.read_csv("베스트셀러_도서구매기록.txt", sep = "\t", engine = "python")
```

```python
df.head()
```

```python
matrix_df = pd.crosstab(index = df['회원번호'], columns = df['책제목'])
matrix_df.head()
```

```python
# 군집화 모델 인스턴스화 및 학습
from sklearn.cluster import AgglomerativeClustering as AC
clustering_model = AC(n_clusters = 10,
                      affinity = "jaccard",
                      linkage = "average")
clustering_model.fit(matrix_df)
```

```python
cluster_labels = clustering_model.labels_
cluster_labels
```

```python
# 회원별 소속 군집 확인
cluster_info = pd.DataFrame({"회원ID":matrix_df.index, "소속군집":cluster_labels})
cluster_info.head()
```

```python
cluster_info['소속군집'].value_counts() # 대다수가 0번 군집에 속함
```

```python
matrix_df_with_cluster_info = pd.merge(matrix_df, cluster_info, left_index = True, right_on = '회원ID')
matrix_df_with_cluster_info.head()
```

```python
matrix_df_with_cluster_info.groupby(['소속군집'])[matrix_df.columns].mean()
```

```python
# 군집별로 가장 많이 구매하는 책 목록 확인
matrix_df_with_cluster_info.groupby(['소속군집'])[matrix_df.columns].mean().idxmax(axis = 1)
```
