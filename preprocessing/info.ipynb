## axis
- 결과 구조 벡터 - (0 : 결과가 행백터, 1: 결과가 열백터)
- 결과 구조 행렬 - (0 : 연산 과정이 행 기준, 1: 연산 과정이 열 기준)

## describe : 열별로 대표적인 기초 통계를 반환

```python
import pandas as pd
```

```python
df = pd.read_csv('온라인_판매기록.csv', encoding='euc-kr') # UnicodeDecodeError
df.head()
```

```python
df[['수량', '판매금액']].mean()
```

```python
df[['수량', '판매금액']].quantile(0.1)
```

```python
df[['수량', '판매금액']].describe()
```

## pivot : 조건에 따른 변수들의 통계량으로 요약
 - index : 행에 들어갈 조건
 - columns : 열에 들어갈 조건
 - values : 집계 대상 컬럼 목록
 - aggfunc : 집계함수

```python
pt1 = pd.pivot_table(df, index = '제품', columns = '쇼핑몰 유형', values = '판매금액', aggfunc = 'max')
pt1
```

```python
pt2 = pd.pivot_table(df, index = '제품', columns = '쇼핑몰 유형', values = ['판매금액','수량'], aggfunc = 'max')
pt2
```
